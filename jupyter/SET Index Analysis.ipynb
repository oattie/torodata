```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../toroapp/static/data/set.csv", index_col="Date", parse_dates=True)
```

```python
df['Adj Close']["2017":"2018"].plot(title="SET Index - www.toro.in.th", figsize=(20, 7))
```

```python
ohlc_dict = {'Open':'first','High':'max','Low':'min','Close': 'last','Volume': 'sum','Adj Close': 'last'}
df_m = df.resample('M', how=ohlc_dict)
```

```python
# Monthly return
m_return = df_m.pct_change(1)*100
m_return.tail(12)
```

```python
plt.figure(figsize=(20, 7))
ax = m_return['Adj Close']["2010":"2019"].plot(title="SET Index - www.toro.in.th")
ax.axhline(y=0, color='r')
```

```python
# Select only Jan of every year
m_jan = m_return.loc[m_return.index.month==1] 
```

```python
plt.figure(figsize=(20, 7))
ax = m_jan['Adj Close'].plot(title="SET Index - www.toro.in.th")   
ax.axhline(y=0, color='r')
ax.axhline(y=m_return['Adj Close'].mean(), color='g')
```
